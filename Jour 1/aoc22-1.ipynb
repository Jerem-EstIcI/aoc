```python
with open('input1a.txt') as file:
    data = [i for i in file.read().strip().split("\n")]

    # faire tout les valeurs dans le fichier texte
    max1 = 0  # 1er
    max2 = 0  # 2eme
    max3 = 0  # 3eme
    count = 0

    for item in data:
        if item == '':
            # Remise a zéro
            if count > max1:
                max3 = max2
                max2 = max1
                max1 = count
            elif count > max2:
                max3 = max2
                max2 = count
            elif count > max3:
                max3 = count
            count = 0
        else:
            num = int(item)
            count += num

    # Check the last group of numbers
    if count > max1:
        max3 = max2
        max2 = max1
        max1 = count
    elif count > max2:
        max3 = max2
        max2 = count
    elif count > max3:
        max3 = count

    print("Total des trois", max1 + max2 + max3)

print("1er", max1)
print("2eme", max2)
print("3eme", max3)
```
