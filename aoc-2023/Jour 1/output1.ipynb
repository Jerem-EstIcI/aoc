```python
with open('input1.txt', 'r') as f:
    lignes = f.readlines()

def chiffres(s):
    """
    >>>chiffres('5ffour295')
    '5295'
    """
    num_dict = {
        'one': '1',
        'two': '2',
        'three': '3',
        'four': '4',
        'five': '5',
        'six': '6',
        'seven': '7',
        'eight': '8',
        'nine': '9'
    }

    ss = ''
    words = s.split()
    found_digit = False

    for word in words:
        if word.isdigit():
            ss += word
            found_digit = True
        elif word in num_dict:
            ss += num_dict[word]
            found_digit = True

    if not found_digit:
        ss = '0'  # Si aucun chiffre ou mot représentant un chiffre n'est trouvé, on met '0'

    return ss


ll = [chiffres(l) for l in lignes]

premetder = [int(s[0])*10 + int(s[-1]) for s in ll]

result_part_two = sum(premetder)
print("Sum of all calibration values (Part Two):", result_part_two)
```
