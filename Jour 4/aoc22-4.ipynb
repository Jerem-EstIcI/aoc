```python
#part1
nombre=0
with open("input4a.txt") as file:
    for ligne in file:
        ligne=ligne.strip("\n").split(",")
        u=ligne[0].split('-')
        a=int(u[0])
        b=int(u[1])
        v=ligne[1].split('-')
        c=int(v[0])
        d=int(v[1])
        if a<=c and d<=b or c<=a and b<=d:
            nombre=nombre+1

print(nombre)
```
