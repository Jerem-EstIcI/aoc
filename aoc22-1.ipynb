```python
#Part 1 
with open ('input1a.txt') as file:
    data = [i for i in file.read().strip().split("\n")]
    
    # faire tout les valeurs dans le fichier texte
    max=0
    count=0
    for item in data:
        if item == '':
            count=0 #Remise a zéro
        else:
            num=int(item)
            count += num
            
        if count > max:
            max = count
            
    print("1er",max)
```

```python
#Part 2
with open ('input1a.txt') as file:
    data = [i for i in file.read().strip().split("\n")]
    
    # faire tout les valeurs dans le fichier texte
    max=0 #1er
    max2=0 #2eme
    max3=0 #3eme
    count=0
    for item in data:
        if item == '':
            count=0 #Remise a zéro
        else:
            num=int(item)
            count += num
            
        if count > max:
            max = count
        elif count > max2:
            max2 = count
        elif count > max3:
            max3 = count
            
    print("Total des trois",max+max2+max3)
print("1er",max)
print("2eme",max2)
print("3eme",max3)
```
