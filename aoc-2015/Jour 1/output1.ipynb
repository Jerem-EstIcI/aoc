```python
def valide(s):
    """
    vérifie que le parenthésage de la chaîne s est correct :
    * autant de ) que de (
    * aucune parenthèse n'est fermée avant d'être ouverte
    >>> valide('(()(()))')
    True
    >>> valide('((()()')
    False
    >>> valide('())(()')
    False
    """
    n = 0  # on part de n=0
    for caractere in s:  # pour chaque caractère dans la chaîne s
        if caractere == '(':  # si c'est une (
            n += 1  # n augmente de 1
        elif caractere == ')':  # si c'est une )
            n -= 1  # n diminue de 1
    return n 
```

```python
def position_sous_sol(s):
    n = 0  # on part de n=0
    for position, caractere in enumerate(s, start=1):  # pour chaque caractère et sa position dans la chaîne s
        if caractere == '(':  # si c'est une (
            n += 1  # n augmente de 1
        elif caractere == ')':  # si c'est une )
            n -= 1  # n diminue de 1
            if n == -1:  # on vérifie si on atteint le sous-sol
                return position  # on renvoie la position du caractère qui fait entrer dans le sous-sol
    return None  # si le sous-sol n'est jamais atteint, on renvoie None
```
