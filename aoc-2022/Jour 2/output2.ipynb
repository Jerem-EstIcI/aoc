```python
#Part 1
score=0
with open('input2a.txt','r') as f:
    for l in f:
        opp=l[0]
        jou=l[2]
        if opp=='A' and jou=='X':
            score=score+4
        if opp=='A' and jou=='Y':
            score=score+8        
        if opp=='A' and jou=='Z':
            score=score+3        
        if opp=='B' and jou=='X':
            score=score+1        
        if opp=='B' and jou=='Y':
            score=score+5        
        if opp=='B' and jou=='Z':
            score=score+9        
        if opp=='C' and jou=='X':
            score=score+7        
        if opp=='C' and jou=='Y':
            score=score+2        
        if opp=='C' and jou=='Z':
            score=score+6
        print(score) #dernier nombre
```
