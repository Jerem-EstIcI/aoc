```python
#part 1
with open('input6a.txt', 'r') as file:
    data = file.read()

def trouver_debut_paquet(data):
    caracteres_traites = 0
    tampon = []

    for caractère in data:
        # Ajouter le caractère au tampon
        tampon.append(caractère)

        # Ajouter le nombre total de caractères traités
        caracteres_traites += 1

        # Si le tampon a plus de 4 caractères, vérifier le marqueur
        if caracteres_traites >= 4:
            if len(set(tampon[-4:])) == 4:
                # marqueur détecté, renvoyer le nombre total de caractères traités
                return caracteres_traites

            # Si le tampon dépasse 4 caractères, retirer le premier caractère du tampon
            tampon.pop(0)

    # Si aucun marqueur n'a pas était trouvé
    return False

# Afficher le résultat
print("Premier marqueur après le caractère:", trouver_debut_paquet(data))
```
